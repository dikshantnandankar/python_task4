```python
student = {
    "id": 101,                          
    "name": "Dikshant Nandankar",              
    "gender": "male",                           
    "course": "data analytics",                 
    "skills": ["Excel", "sql", "power bi", "python"],  
    "graduated": "yes",               
    "address": {                         
        "city": "Nagpur",
        "state": "Maharashtra",
        "pincode": 440008
    },
    "phone": 8446568570
}

print(f"student = {student}, type of student = {type(student)}, length of student = {len(student)}")
print('-'*10)
print(f"student skills = {student['skills']}, type of student skills = {type(student['skills'])}, length of student skills = {len(student['skills'])}")
print('-'*10)
print(f"student address = {student['address']}, type of student address= {type(student['address'])}, length of student address = {len(student['address'])}")

```

**clear**

```python
print(f"student after clear = {student.clear()}")
```

**copy**

```python
student_data = student.copy()
print(f"student_update after copy = {student_data}")
```

**get**

```python
print(f"get student skills = {student.get("skills")}")
print('-'*10)
student["skills"].append("alteryx")
print(f"get student skills after append = {student.get("skills")}")

```

**items**

```python
print(f"items of student = {student.items()}")
```

**keys**

```python
print(f"keys of student = {student.keys()}")
```

**setdefault**

```python
print(student)  
print(f"setdefault of student key course = {student.setdefault("course")}")
```

**update**

```python
student.update({"department" : "IT"})

print(f"after update student = {student}")
```

**values**

```python
print(f"values of student = {student.values()}")
student["graduated"] = "no"
print('-'*10)
print(f"after changing graduated values of student = {student.values()}")
```

**popitem**

```python
student.popitem()
print(f"after popitem student = {student}")
```
