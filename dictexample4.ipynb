```python
employee = {
    "emp_id" : 101,
    "name" : "dikshant",
    "department" : "IT",
    "salary" : 20000,
    "city" : "mumbai"
}

print(f"employee = {employee}, length of employee = {len(employee)}, type of employee = {type(employee)}")
```

**clear**

```python
print(f"employee after clear = {employee.clear()}")
```

**copy**

```python
employee_details = employee.copy()
print(f"employee_details after copy = {employee_details}")
```

**get**

```python
print(f"get employee city = {employee.get("city")}")
```

**items**

```python
print(f"items of employee = {employee.items()}")
```

**keys**

```python
print(f"keys of employee = {employee.keys()}")
```

**setdefault**

```python
print(employee)  
print(f"setdefault of employee key name = {employee.setdefault("name")}")
```

**update**

```python
employee.update({"state" : "maharashtra"})

print(f"after update employee = {employee}")
```

**values**

```python
print(f"values of employee = {employee.values()}")
employee["city"] = "nagpur"
print('-'*10)
print(f"after changing city values of employee = {employee.values()}")
```

**pop**

```python
print(f"pop of employee = {employee.pop("name")}")
```

```python
print(employee)
```

**popitem**

```python
employee.popitem()
print(f"after popitem employee = {employee}")
```
