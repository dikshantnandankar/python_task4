```python
laptop_entry = {
    "brand": "acer",
    "model": "aspire lite",
    "year": 2026,
    "processor": "Intel Core i5",
    "graphics_card": "NVIDIA",
    "ram_gb": 16,
    "storage_capacity_gb": 500,
    "display_type": "OLED",
    "refresh_rate_hz": 240,
    "resolution": "3840x2400",
    "battery_wh": 95,
    "weight_kg": 2.45,
    "os": "Windows 11",
    "warranty_years": 1,
    "price": 40000
}

print(f"laptop_entry = {laptop_entry}, length of laptop_entry = {len(laptop_entry)}, type of laptop_entry = {type(laptop_entry)}")
```

**clear**

```python
print(f"laptop_entry after clear = {laptop_entry.clear()}")
```

**copy**

```python
laptop_entry_update = laptop_entry.copy()
print(f"laptop_entry_update after copy = {laptop_entry_update}")
```

**get**

```python
print(f"get laptop_entry model = {laptop_entry.get("model")}")
```

**items**

```python
print(f"items of laptop_entry = {laptop_entry.items()}")
```

**keys**

```python
print(f"keys of laptop_entry = {laptop_entry.keys()}")
```

**setdefault**

```python
print(laptop_entry)  
print(f"setdefault of laptop_entry key processor = {laptop_entry.setdefault("processor")}")
```

**update**

```python
laptop_entry.update({"stock" : 10})

print(f"after update laptop_entry = {laptop_entry}")
```

**values**

```python
print(f"values of laptop_entry = {laptop_entry.values()}")
laptop_entry["ram_gb"] = 32
print('-'*10)
print(f"after changing ram_gb values of laptop_entry = {laptop_entry.values()}")
```

**popitem**

```python
laptop_entry.popitem()
print(f"after popitem laptop_entry = {laptop_entry}")
```
