```python
phone = {
    "brand" : "Apple",
    "model" : "Iphone 16 pro",
    "storage_gb" : 256,
    "color" : "gold",
    "price" : 100000
}

print(f"phone = {phone}, length of phone = {len(phone)}, type of phone = {type(phone)}")
```

**clear**

```python
print(f"phone after clear = {phone.clear()}")
```

**copy**

```python
phone_data = phone.copy()
print(f"phone update after copy = {phone_data}")
```

**get**

```python
print(f"get phone model = {phone.get("model")}")
```

**items**

```python
print(f"items of phone = {phone.items()}")
```

**keys**

```python
print(f"keys of phone = {phone.keys()}")
```

**setdefault**

```python
print(phone)  
print(f"setdefault of phone key model = {phone.setdefault("model")}")
```

**update**

```python
phone.update({"model_year" : 2025})

print(f"after update phone = {phone}")
```

**values**

```python
print(f"values of phone = {phone.values()}")
phone["model"] = "Iphone 17 pro"
print('-'*10)
print(f"after changing model values of phone = {phone.values()}")
```

**popitem**

```python
phone.popitem()
print(f"after popitem phone = {phone}")
```
