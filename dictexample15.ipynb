```python
machine = {
    "machine_id" : 1001,
    "name" : "cutter machine",
    "type" : "automatic",
    "manufacture" : "bosch",
    "price" : 10000
}
print(f"machine = {machine}, length of machine = {len(machine)}, type of machine ={type(machine)}")
```

**clear**

```python
print(f"machine after clear = {machine.clear()}")
```

**copy**

```python
machine_data = machine.copy()
print(f"copy of machine = {machine_data} ")
```

**get**

```python
print(f"get machine type = {machine.get("type")}")
```

**items**

```python
print(f"items of machine = {machine.items()}")
```

**keys**

```python
print(f"keys of machine = {machine.keys()}")
```

**setdefault**

```python
print(machine)
print(f"setdefault machine key price = {machine.setdefault("price")}")
```

**update**

```python
machine.update({"country" : "India"})
print(f"machine after update = {machine}")
```

**values**

```python
print(f"values of machine = {machine.values()}")
machine["price"] = 20000
print('-'*10)
print(f"after changing price values of machine = {machine.values()}")
```

**pop**

```python
print(f"pop of machine = {machine.pop("machine_id")}")
```

```python
print(machine)
```

**popitems**

```python
machine.popitem()
print(f"after popitem machine = {machine}")
```
