```python
car = {
    "brand": "tata",
    "model": "safari",
    "year": 2025,
    "features": ["Airbags", "ABS", "Sunroof"],  
    "engine": {                                 
        "type": "Diesel",
        "capacity": "2898cc",
        "transmission" : "automatic"
    }
}

print(f"car = {car}, length of car = {len(car)}, type of car = {type(car)}")
print('-'*10)
print(f"car details of engine = {car['engine']}, length of engine = {len(car['engine'])}, type of engine = {type(car['engine'])}")
print('-'*10)
print(f"car details of features = {car['features'][2]}, length of features = {len(car['features'])}, type of features = {type(car['features'])}")
```

**clear**

```python
print(f"car after clear = {car.clear()}")
```

**copy**

```python
car_data = car.copy()
print(f"copy of car = {car_data} ")
```

**get**

```python
print(f"get car year = {car.get("year")}")
```

**items**

```python
print(f"items of car = {car.items()}")
```

**keys**

```python
print(f"keys of car = {car.keys()}")
```

**setdefault**

```python
print(car)
print(f"setdefault car key features = {car.setdefault("features")}")
```

**update**

```python
car.update({"price" : 3000000})
print(f"car after update = {car}")
```

**values**

```python
print(f"values of car = {car.values()}")
car["year"] = 2026
print('-'*10)
print(f"after changing year values of car = {car.values()}")
```

**popitem**

```python
car.popitem()
print(f"after popitem car = {car}")
```
