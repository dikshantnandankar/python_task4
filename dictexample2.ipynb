```python
product = {
    "product_id" : 101,
    "name" : "laptop",
    "brand" : "acer",
    "details" : {
        "RAM" : "16gb",
        "graphics_card" : "2gb"
    },
    "price" : 44000
}
print(f"product = {product}, length of product = {len(product)}, type of product = {type(product)}")
```

**clear**

```python
print(f"product after clear = {product.clear()}")
```

**copy**

```python
product_details = product.copy()
print(f"product_details after copy = {product_details}")
```

**get**

```python
print(f"get product details = {product.get("details")}")
```

**items**

```python
print(f"items of product = {product.items()}")
```

**keys**

```python
print(f"keys of product = {product.keys()}")
```

**setdefault**

```python
print(product)
print(f"setdefault of product key brand = {product.setdefault("brand")}")
```

**update**

```python
product.update({"stock" : 10})

print(f"after update product = {product}")
```

**values**

```python
print(f"values of product = {product.values()}")
product["product_id"] = 102
print('-'*10)
print(f"values of product_id after changing = {product.values()}")
```

**pop**

```python
print(f"pop of product = {product.pop("brand")}")
```

```python
print(product)
```

**popitem**

```python
product.popitem()
print(f"after popitem product = {product}")
```
