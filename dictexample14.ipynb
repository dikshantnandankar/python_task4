```python
order = {
    "drink" : "Cold coffee", 
    "size" : "large",
    "price" : 200,
    "payment_mode" : "cash",
    "order_no" : "cc101"
}

print(f"order = {order}, length of order = {len(order)}, type of order = {type(order)}")
```

**clear**

```python
print(f"order after clear = {order.clear()}")
```

**copy**

```python
order_data = order.copy()
print(f"copy of order = {order_data} ")
```

**get**

```python
print(f"get order price = {order.get("price")}")
```

**items**

```python
print(f"items of order = {order.items()}")
```

**keys**

```python
print(f"keys of order = {order.keys()}")
```

**setdefault**

```python
print(order)
print(f"setdefault order key size = {order.setdefault("size")}")
```

**update**

```python
order.update({"order_time" : "10 minutes"})
print(f"order after update = {order}")
print(f"type of order_time = {type(order['order_time'])}")
```

**values**

```python
print(f"values of order = {order.values()}")
order["size"] = "medium"
print('-'*10)
print(f"after changing size values of order = {order.values()}")
```

**popitems**

```python
order.popitem()
print(f"after popitem order = {order}")
```
