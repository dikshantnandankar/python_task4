```python
mobile = {
    "brand": "Samsung",
    "model": "Galaxy S24",
    "price": 75000,
    "color": "Black",
    "ram": "8GB",
    "storage": "256GB",
    "battery": "3900mAh",
    "display": "6 inch",
    "camera_rear": "50MP",
    "camera_front": "12MP",
    "processor": "Snapdragon 8",
    "os": "Android",
    "fingerprint": "Yes",
    "5g_support": True,
    "release_year": 2024
}

print(f"mobile = {mobile}, length of mobile = {len(mobile)}, type of mobile = {type(mobile)}")
```

**clear**

```python
print(f"mobile after clear = {mobile.clear()}")
```

**copy**

```python
mobile_data = mobile.copy()
print(f"copy of mobile = {mobile_data} ")
```

**get**

```python
print(f"get mobile color = {mobile.get("color")}")
```

**items**

```python
print(f"items of mobile = {mobile.items()}")
print(f"items of mobile brand = {mobile.items()}")
```

**keys**

```python
print(f"keys of mobile = {mobile.keys()}")
```

**setdefault**

```python
print(mobile)
print(f"setdefault mobile key price = {mobile.setdefault("price")}")
```

**update**


```python
mobile.update({"Charger" : "yes"})
print(f"mobile after update = {mobile}")
```

**values**

```python
print(f"values of mobile = {mobile.values()}")
mobile["color"] = "silver"
print('-'*10)
print(f"after changing color values of mobile = {mobile.values()}")
```

**popitem**

```python
mobile.popitem()
print(f"after popitem mobile = {mobile}")
```
