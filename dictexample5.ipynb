```python
employee_data = {
    "emp_id": 101,
    "first_name": "dikshant",
    "last_name": "nandankar",
    "department": "IT",
    "role": "junior data Analyst",
    "manager": "someone",
    "hire_date": "14-03-2026",
    "salary": 20000,
    "remote_status": "Hybrid",
    "office_location": "Mumbai",
    "email": "dikshant@gmail.com",
    "phone_no": 8446568570,
    "certifications": ["data analyst", "data scientist"],
    "performance_rating": 4.8,
    "is_active": True,
}

print(f"employee_data = {employee_data}, length of employee_data = {len(employee_data)}, type of employee_data = {type(employee_data)}")
```

**clear**

```python
print(f"employee data after clear = {employee_data.clear()}")
```

**copy**

```python
employee_details = employee_data.copy()
print(f"employee_details after copy = {employee_details}")
```

**get**

```python
print(f"get employee_data role = {employee_data.get("role")}")
```

**items**

```python
print(f"items of employee_dat = {employee_data.items()}")
```

**keys**

```python
print(f"keys of employee_data = {employee_data.keys()}")
```

**setdefault**

```python
print(employee_data)  
print(f"setdefault of employee_data key hire_date = {employee_data.setdefault("hire_date")}")
```

**update**

```python
employee_data.update({"time" : 900})

print(f"after update employee_data = {employee_data}")
```

**values**

```python
print(f"values of employee_data = {employee_data.values()}")
employee_data["office_location"] = "nagpur"
print('-'*10)
print(f"after changing office_location values of employee_data = {employee_data.values()}")
```

**pop**

```python
print(f"pop of employee_data = {employee_data.pop("emp_id")}")
```
