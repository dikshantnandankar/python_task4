```python
car = {
    "brand" : "hyundai",
    "model" : "venue",
    "year" : 2023,
    "color" : ["white", "black", "grey", "blue"],
    "price" : 1200000
}

print(f"car = {car}, length of car = {len(car)}, type of car = {type(car)}")
```

**clear**

```python
print(f"car after clear = {car.clear()}")
```

**copy**

```python
car_details = car.copy()
print(f"car_details after copy = {car_details}")
```

**get**

```python
print(f"get car color = {car.get("color")}")
```

**items**

```python
print(f"items of car = {car.items()}")
```

**keys**

```python
print(f"keys of car = {car.keys()}")
```

**setdefault**

```python
print(car)  
print(f"setdefault of car key model = {car.setdefault("model")}")
```

**update**

```python
car.update({"price" : 1300000})

print(f"after update car = {car}")
```

**values**

```python
print(f"values of car = {car.values()}")
car["year"] = 2024
print('-'*10)
print(f"after changing year values of car = {car.values()}")
```

**pop**

```python
print(f"pop of car = {car.pop("year")}")
```

**popitem**

```python
car.popitem()
print(f"after popitem car = {car}")
```
