```python
weather = {
    "City" : "nagpur",
    "temperature" : 40.7,
    "day" : "sunday",
    "condition" : "sunny",
    "wind_speed" : 20
}

print(f"weather = {weather}, length of weather = {len(weather)}, type of weather = {type(weather)}")
```

**clear**

```python
print(f"weather after clear = {weather.clear()}")
```

**copy**

```python
weather_data = weather.copy()
print(f"copy of weather = {weather_data} ")
```

**get**

```python
print(f"get weather City = {weather.get("City")}")
```

**items**

```python
print(f"items of weather = {weather.items()}")
```

**keys**

```python
print(f"keys of weather = {weather.keys()}")
```

**setdefault**

```python
print(weather)
print(f"setdefault weather key day = {weather.setdefault("day")}")
```

**update**

```python
weather.update({"humidity" : 13})
print(f"weather after update = {weather}")
```

**values**

```python
print(f"values of weather = {weather.values()}")
weather["City"] = "mumbai"
print('-'*10)
print(f"after changing City values of weather = {weather.values()}")
```

**popitems**

```python
weather.popitem()
print(f"after popitem weather = {weather}")
```
