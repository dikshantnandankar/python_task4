```python
car = {
    "brand": "Toyota",
    "model": "Fortuner",
    "year": 2025,
    "color": "white",
    "fuel_type": "Diesel",
    "transmission": "Automatic",
    "engine": "2898cc",
    "mileage": "10 km/l",
    "price": 4800000,
    "seats": 7,
    "airbags": 6,
    "sunroof": "false",
    "navigation": "True",
    "warranty": "3 years",
    "country": "Japan"
}

print(f"car = {car}, length of car = {len(car)}, type of car = {type(car)}")
```

**clear**

```python
print(f"car data after clear = {car.clear()}")
```

**copy**

```python
car_data = car.copy()
print(f"car_data after copy = {car}")
```

**get**

```python
print(f"get car color = {car.get("color")}")
```

**items**

```python
print(f"items of car = {car.items()}")
```

**keys**

```python
print(f"keys of car = {car.keys()}")
```

**setdefault**

```python
print(car)  
print(f"setdefault of car key price = {car.setdefault("price")}")
```

**update**

```python
car.update({"screen" : "15 inch"})

print(f"after update car = {car}")
```

**values**

```python
print(f"values of car = {car.values()}")
car["color"] = "black"
print('-'*10)
print(f"after changing color values of car = {car.values()}")
```

**popitem**

```python
car.popitem()
print(f"after popitem car = {car}")
```
