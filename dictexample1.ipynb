```python
student = {
    "student_id" : 101,
    "name" : "Dikshant",
    "course" : "data analytics",
    "topis" : ["excel", "sql", "power bi", "python"],
    "city" : "nagpur"
}

print(f"student = {student}, length of student = {len(student)}, type of student = {type(student)}")
```

**clear** => **removes all elements from dictionary**

```python
print(f"student after clear = {student.clear()}")
```

```python
print(student)
```

**copy** => **returns copy of dictionary**

```python
student_details = student.copy()
print(f"student_details after copy = {student_details}")
```

**get** => **returns specified key from dictionary**

```python
print(f"get student topics = {student.get("topis")}")
```

**items** => **contains key and values of dictionary**


```python
print(f"items of student = {student.items()}")
```

**keys** => **returns list of keys in dictionary**

```python
print(f"keys of student = {student.keys()}")
```

**pop** => **removes element of specified key**

```python
print(f"pop of student = {student.pop("city")}")

```

```python
print(student)
```

**setdefault** => **returns value of specified key**

```python
print(student)
print(f"setdefault of student key course = {student.setdefault("course")}")
```

**update** => **update dictionary with specified key value pair**

```python
student.update({"state" : "maharashtra"})

print(f"after update student = {student}")
```

**values** => **returns list of all values in dictionary**

```python
print(f"values of student = {student.values()}")
student["student_id"] = 102
print('-'*10)
print(f"values of student_id after changing = {student.values()}")
```

**popitem** => **removes last inserted key value pair**

```python
student.popitem()
print(f"after popitem student = {student}")
```
