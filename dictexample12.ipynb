```python
student = {
    "name": "Dikshant",
    "email": "dikshant@gmail.com",
    "course": "data analytics",
    "subjects": ["Excel", "sql", "power bi" , "Python"],   
    "status": {
        "excel" : "done",
        "sql" : "done",
        "power_bi" :"done",                                
        "Python": "ongoing"
    }
}

print(f"student = {student}, length of student = {len(student)}, type of student = {type(student)}")
```

**clear**

```python
print(f"student after clear = {student.clear()}")
```

**copy**

```python
student_data = student.copy()
print(f"copy of student = {student_data} ")
```

**get**

```python
print(f"get student course = {student.get("course")}")
```

**items**

```python
print(f"items of student = {student.items()}")
```

**keys**

```python
print(f"keys of student = {student.keys()}")
```

**setdefault**

```python
print(student)
print(f"setdefault student key subjects = {student.setdefault("subjects")}")
```

**update**

```python
student.update({"id" : 103})
print(f"student after update = {student}")
```

**values**

```python
print(f"values of student = {student.values()}")
student["course"] = "data science"
print('-'*10)
print(f"after changing course values of student = {student.values()}")
```

**popitems**

```python
student.popitem()
print(f"after popitem student = {student}")
```
